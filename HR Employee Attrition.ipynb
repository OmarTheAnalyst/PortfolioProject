```python
df = pd.read_csv('HR Employee Attrition.csv')
```

```python
#check null values
for col in df.columns:
    df_missing = np.count_nonzero(df[col].isnull())
    print ('{} -{}'.format(col, df_missing))
```

```python
plt.figure(figsize=(8,6))
plt.pie(Attrition_count, labels=['No','Yes'],explode= (0.1,0),autopct='%.1f%%', wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'},textprops={'size': 'x-large'})
plt.title('Employee Attrition',fontsize=15)
plt.tight_layout()
```
